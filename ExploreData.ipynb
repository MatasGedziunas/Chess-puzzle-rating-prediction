```python
import pandas as pd
```

```python
data = pd.read_csv("./data/p200k.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the plot is large enough to read
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

# Determine the min and max to create clean bins of 100
min_rating = int(data['Rating'].min() // 100 * 100)
max_rating = int(data['Rating'].max() // 100 * 100) + 100
bins = range(min_rating, max_rating + 100, 100)

# Plot the histogram
sns.histplot(data=data, x='Rating', bins=bins, color='steelblue', edgecolor='black')

# Formatting
plt.title('Distribution of Puzzle Ratings', fontsize=16)
plt.xlabel('Rating (Buckets of 100)', fontsize=14)
plt.ylabel('Count of Puzzles', fontsize=14)

# Optional: Add tick marks every 200 rating points for better readability
plt.xticks(range(min_rating, max_rating + 100, 200)) 

plt.show()
```

```python
count_low = len(data[data['Rating'] < 1200])
count_mid = len(data[(data['Rating'] >= 1200) & (data['Rating'] <= 1800)])
count_high = len(data[data['Rating'] > 1800])

print(f"Puzzles < 1200: {count_low:,}")
print(f"Puzzles 1200 - 1800: {count_mid:,}")
print(f"Puzzles > 1800: {count_high:,}")
```
