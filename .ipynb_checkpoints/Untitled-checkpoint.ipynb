```python
import pandas as pd

data = pd.read_csv("data/data.csv")

TARGET = "Rating"

FEATURES = [c for c in data.columns if c.startswith("success_prob_")]

X = data[FEATURES]
y = data[TARGET]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.1, random_state=42
)
```

```python
import lightgbm as lgb

model = lgb.LGBMRegressor(
    objective="regression",
    n_estimators=1000,
    learning_rate=0.05,
    num_leaves=64,
    max_depth=-1,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)

model.fit(
    X_train,
    y_train,
    eval_set=[(X_val, y_val)],
    eval_metric="mse",
    callbacks=[
        lgb.early_stopping(stopping_rounds=50),
        lgb.log_evaluation(50)
    ]
)
```

```python
import joblib

joblib.dump(model, "lgbm_puzzle_rating_model.joblib")
```

```python
import joblib
model = joblib.load("lgbm_puzzle_rating_model.joblib")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import chess
import chess.svg
from IPython.display import SVG, display
from sklearn.metrics import mean_squared_error

# --- Pick a random validation sample ---
idx = np.random.choice(X_val.index)
sample = data.loc[idx]

fen = sample["FEN"]
true_rating = sample["Rating"]
pred_rating = model.predict(X_val.loc[[idx]])[0]
error = pred_rating - true_rating

# --- Render chess board ---
board = chess.Board(fen)
display(SVG(chess.svg.board(board=board, size=420)))

# --- Print prediction info ---
print(f"🧩 Puzzle ID: {sample['PuzzleId']}")
print(f"⭐ Actual Rating:     {true_rating:.0f}")
print(f"🤖 Predicted Rating: {pred_rating:.0f}")
print(f"📉 Error:            {error:+.0f}")
print()
print("♟ Solution moves:")
print(sample["Moves"])

# --- Plot success probability curves ---
ratings = [1050,1150,1250,1350,1450,1550,1650,1750,1850,1950,2050]

rapid_probs = [sample[f"success_prob_rapid_{r}"] for r in ratings]
blitz_probs = [sample[f"success_prob_blitz_{r}"] for r in ratings]

plt.figure(figsize=(6,4))
plt.plot(ratings, rapid_probs, marker="o", label="Rapid")
plt.plot(ratings, blitz_probs, marker="o", label="Blitz")
plt.axvline(true_rating, linestyle="--", label="True rating")
plt.axvline(pred_rating, linestyle=":", label="Predicted rating")
plt.xlabel("Player rating")
plt.ylabel("Success probability")
plt.title("Success probability curve")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

val_preds = model.predict(X_val)
mse = mean_squared_error(y_val, val_preds)

mse
```

```python
final_model = lgb.LGBMRegressor(
    objective="regression",
    n_estimators=model.best_iteration_,
    learning_rate=0.05,
    num_leaves=64,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)

final_model.fit(X, y)
```
