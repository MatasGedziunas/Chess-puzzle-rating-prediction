```python
from onnx2torch import convert
import torch

model = convert("maia2_models/maia1100.onnx")
model.eval()

# Use a forward hook to grab the intermediate output
embeddings = {}
def hook_fn(module, input, output):
    embeddings['emb'] = output

# Register hook on the layer BEFORE the heads
handle = getattr(model, "block5/conv2/mixin").register_forward_hook(hook_fn)

# Run a dummy forward pass
dummy_input = torch.randn(1, 112, 8, 8)
with torch.no_grad():
    _ = model(dummy_input)

# Now embeddings['emb'] has the [1, 64, 8, 8] tensor
emb = embeddings['emb'].flatten(1)  # → [1, 4096]
handle.remove()
```

```python
import torch
import numpy as np
import chess
from onnx2torch import convert
from tqdm import tqdm

NUM_MOVES = 5  # number of board states to extract (initial + moves)


def fen_to_lc0_planes(fen: str) -> np.ndarray:
    """Converts a FEN string to a 112x8x8 input tensor for Maia/Lc0."""
    board = chess.Board(fen)
    planes = np.zeros((112, 8, 8), dtype=np.float32)
    stm = board.turn

    def get_piece_mask(color, piece_type):
        mask = np.zeros(64, dtype=np.float32)
        for sq in board.pieces(piece_type, color):
            mask[sq ^ 56] = 1.0
        return mask.reshape((8, 8))

    pieces = [chess.PAWN, chess.KNIGHT, chess.BISHOP, chess.ROOK, chess.QUEEN, chess.KING]
    for i, p in enumerate(pieces):
        planes[i] = get_piece_mask(stm, p)
        planes[i + 6] = get_piece_mask(not stm, p)

    if stm == chess.BLACK:
        planes[104].fill(1.0)
    planes[105].fill(board.fullmove_number / 100.0)
    if board.has_kingside_castling_rights(stm): planes[106].fill(1.0)
    if board.has_queenside_castling_rights(stm): planes[107].fill(1.0)
    if board.has_kingside_castling_rights(not stm): planes[108].fill(1.0)
    if board.has_queenside_castling_rights(not stm): planes[109].fill(1.0)
    planes[110].fill(board.halfmove_clock / 100.0)
    planes[111].fill(1.0)
    return planes


def get_puzzle_boards(fen: str, moves_uci: str, num_positions: int = NUM_MOVES) -> list[np.ndarray]:
    """
    Given a FEN and a UCI move string, returns lc0 planes for up to 
    num_positions board states (initial position + after each move).
    Pads with zeros if the puzzle has fewer moves.
    """
    board = chess.Board(fen)
    moves = moves_uci.split()
    
    planes_list = []
    # Initial position
    planes_list.append(fen_to_lc0_planes(board.fen()))
    
    # Apply moves one by one
    for move_uci in moves[:num_positions - 1]:
        board.push_uci(move_uci)
        planes_list.append(fen_to_lc0_planes(board.fen()))
    
    # Pad with zeros if fewer moves than num_positions
    while len(planes_list) < num_positions:
        planes_list.append(np.zeros((112, 8, 8), dtype=np.float32))
    
    return planes_list


class Maia1FeatureExtractor:
    def __init__(self, model_path):
        self.model = convert(model_path)
        self.model.eval()
        self.latent_vector = None
        self.hook = getattr(self.model, "block5/conv2/mixin").register_forward_hook(self._save_output)

    def _save_output(self, module, input, output):
        self.latent_vector = output

    def __call__(self, boards):
        self.model(boards)
        return self.latent_vector.flatten(1)  # [batch, 4096]


def process_dataset_maia1(df, model_paths, num_positions=NUM_MOVES, batch_size=64):
    """
    Extracts embeddings for the first `num_positions` board states of each puzzle.
    
    Returns: np.array of shape (num_puzzles, num_models, num_positions, 4096)
    """
    device = "cuda" if torch.cuda.is_available() else "cpu"

    extractors = {}
    for name, path in model_paths.items():
        ext = Maia1FeatureExtractor(path)
        ext.model.to(device)
        extractors[name] = ext

    fens = df['FEN'].tolist()
    moves = df['Moves'].tolist()
    num_puzzles = len(fens)
    num_models = len(extractors)

    # Shape: (puzzles, models, positions, 4096)
    all_results = np.zeros((num_puzzles, num_models, num_positions, 4096), dtype=np.float32)

    for i in tqdm(range(0, num_puzzles, batch_size)):
        batch_end = min(i + batch_size, num_puzzles)
        batch_size_actual = batch_end - i
        
        # Get all board states for this batch
        # Shape: (batch, num_positions, 112, 8, 8)
        batch_boards = [
            get_puzzle_boards(fens[j], moves[j], num_positions)
            for j in range(i, batch_end)
        ]
        
        for model_idx, (name, extractor) in enumerate(extractors.items()):
            # Process each position across the batch
            for pos_idx in range(num_positions):
                # Stack position pos_idx from all puzzles in batch → [batch, 112, 8, 8]
                pos_planes = np.array([batch_boards[b][pos_idx] for b in range(batch_size_actual)])
                boards_tensor = torch.from_numpy(pos_planes).float().to(device)
                
                with torch.no_grad():
                    latent = extractor(boards_tensor)  # [batch, 4096]
                    all_results[i:batch_end, model_idx, pos_idx, :] = latent.cpu().numpy()

    return all_results
```

```python
model_paths = {
    'maia1100': 'maia2_models/maia1100.onnx',
    'maia1500': 'maia2_models/maia1500.onnx',
    'maia1900': 'maia2_models/maia1900.onnx',
}
df = pd.read_csv("../data/puzzle50000.csv")
results = process_dataset_maia1(df, model_paths)
np.save("../features/maia1_embeddings_sequence5.npy", results)
```
